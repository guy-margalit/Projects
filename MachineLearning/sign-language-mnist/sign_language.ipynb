```python
import numpy as np
import pandas as pd
from PIL import Image
```

```python
df = pd.read_csv("sign_mnist_train.csv")
```

```python
 def to_image(i):
    array = []
    for y in range(1, 29):
        line = []
        for x in range(1, 29):
            line.append(df['pixel' + str(y*x)][i])
        array.append(line)
    array = np.array(array, dtype=np.uint8)
    return Image.fromarray(array)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
pixels = ['pixel' + str(i) for i in range(1, 785)]
```

```python
pixels
```

```python
x = df[pixels]
```

```python
y = df['label']
```

```python
model.fit(x, y)
```

```python
test = pd.read_csv("sign_mnist_test.csv")
```

```python
x_test = test[pixels]
```

```python
y_test = test['label']
```

```python
model.score(x_test, y_test)
```
